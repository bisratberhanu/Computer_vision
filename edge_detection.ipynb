```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("DATA/sammy_face.jpg")
```

```python
edges = cv2.Canny(img, threshold1=127, threshold2=255)
# print(edges.shape)
```

```python
plt.imshow(edges)
```

for better performance using a predetermined formula will help

```python
mid_val = np.median(img)
print(mid_val)
```

```python
# FOR LOWER BOUND, TAKE THE MAX OF 0 AND 70% OF THE MEDIAN VALUE
lower = max(0, 0.7*mid_val) 
upper = min(255, 1.3 * mid_val)
```

```python
edges = cv2.Canny(img,threshold1=lower, threshold2=upper)
plt.imshow(edges) # for this case image doesnt does show better performance
```

```python
# blurring image will help  
blurred_img = cv2.blur(img,ksize=(5,5))
edges = cv2.Canny(blurred_img,threshold1=lower, threshold2=upper+50) #change the upper value to see a bet result(play_around)
plt.imshow(edges)
```
