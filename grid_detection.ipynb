```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
flat_chess = cv2.imread("DATA/flat_chessboard.png")
```

```python
plt.imshow(flat_chess)
```

```python
found,corners = cv2.findChessboardCorners(flat_chess,(7,7)) # 7,7 because last edge is not detected 
```

```python
found
```

```python
corners  # the coordinate points where a corner was found

cv2.drawChessboardCorners(flat_chess,(7,7),corners, found)
```

```python

plt.imshow(flat_chess)
```

grid can also be identified on circular materials

```python
dots = cv2.imread("DATA/dot_grid.png")
plt.imshow(dots)
```

```python
found,corners =  cv2.findCirclesGrid(dots,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
cv2.drawChessboardCorners(dots,(10,10),corners, found)
```
